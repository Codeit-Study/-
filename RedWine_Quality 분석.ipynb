- fixed acidity: 고정 산도
- volatile acidity: 휘발성 산도
- citric acid: 구연산 함량
- residual sugar: 잔류 당분
- chlorides: 염화물 농도(와인의 소금기)
- free sulfur dioxide: 자유 이산화황농도 (와인의 보존제역할)
- total sulfur dioxide: 총 이산화황 농도
- sulphates: 황산염 농도

```python
# 품질(Quality) 분포 시각화
sns.countplot(x='quality', data=wine, palette='viridis')
plt.title("Quality Distribution")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
# 상관 관계 히트맵
plt.figure(figsize=(12, 10))
correlation_matrix = wine.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

와인의 품질(quality)은 alcohol(양의 상관), volatile acidity(음의 상관)와 가장 큰 관계를 가짐.

```python
# 각 feature와 품질(Quality)의 관계
for column in wine.columns[:-1]:  # 'quality' 제외
    sns.boxplot(x='quality', y=column, data=wine, palette='Set3')
    plt.title(f"Quality vs {column}")
    plt.xlabel("Quality")
    plt.ylabel(column)
    plt.show()
```
